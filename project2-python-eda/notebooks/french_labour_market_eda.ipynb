```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (12, 6)
sns.set_palette('husl')
print('Libraries loaded')
```

## Data Loading (based on real INSEE/Eurostat figures)

```python
data_national = {
    'year': list(range(2010, 2024)),
    'rate': [9.3,9.6,10.0,10.4,10.3,10.4,10.1,9.4,9.0,8.5,7.9,8.1,7.3,7.3]
}
data_gender = {
    'year': list(range(2010, 2024)),
    'men':   [9.0,9.8,10.0,10.2,10.1,10.3,10.0,9.2,8.7,8.3,7.7,8.0,7.1,7.0],
    'women': [9.7,9.5,10.1,10.6,10.5,10.6,10.2,9.7,9.4,8.8,8.2,8.3,7.5,7.6]
}
data_region = {
    'region': ['Hauts-de-France','Occitanie','PACA','Normandie','Grand Est','IDF','Nouvelle-Aquitaine','Bourgogne-FC','Auvergne-RA','Bretagne'],
    'rate_2023': [11.3,9.2,8.4,7.2,7.9,7.1,6.8,6.9,6.4,5.8]
}
data_age = {
    'group': ['15-24 ans','25-49 ans','50 ans+'],
    '2015':[24.7,9.8,7.2], '2018':[20.8,8.7,6.8], '2021':[19.3,7.9,6.2], '2023':[17.4,7.1,5.8]
}
df_nat = pd.DataFrame(data_national)
df_gen = pd.DataFrame(data_gender)
df_reg = pd.DataFrame(data_region)
df_age = pd.DataFrame(data_age)
print('Data ready')
df_nat.tail()
```

## Chart 1 — National Unemployment Trend (2010-2023)

```python
fig, ax = plt.subplots(figsize=(13,6))
ax.plot(df_nat['year'], df_nat['rate'], color='#2E86AB', lw=2.5, marker='o', ms=6)
ax.fill_between(df_nat['year'], df_nat['rate'], alpha=0.15, color='#2E86AB')
ax.axvline(x=2020, color='red', ls='--', alpha=0.6, label='COVID-19')
ax.set_title('French Unemployment Rate (2010-2023)', fontsize=16, fontweight='bold')
ax.set_xlabel('Year'); ax.set_ylabel('Rate (%)')
ax.legend(); ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../dashboard/chart1_trend.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'Min: {df_nat.rate.min()}% | Max: {df_nat.rate.max()}%')
```

## Chart 2 — Unemployment by Age Group

```python
years=['2015','2018','2021','2023']
x=np.arange(len(years)); width=0.25
colors=['#E84855','#3A86FF','#06D6A0']
fig,ax=plt.subplots(figsize=(12,6))
for i,(_, row) in enumerate(df_age.iterrows()):
    vals=[row[y] for y in years]
    bars=ax.bar(x+i*width, vals, width, label=row['group'], color=colors[i], alpha=0.85)
    for b,v in zip(bars,vals):
        ax.text(b.get_x()+b.get_width()/2, b.get_height()+0.3, f'{v}%', ha='center', fontsize=9, fontweight='bold')
ax.set_title('Unemployment by Age Group — France', fontsize=16, fontweight='bold')
ax.set_xticks(x+width); ax.set_xticklabels(years)
ax.legend(title='Age group'); ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../dashboard/chart2_age.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Chart 3 — Gender Gap

```python
fig,ax=plt.subplots(figsize=(13,6))
ax.plot(df_gen['year'], df_gen['men'], color='#3A86FF', lw=2.5, marker='o', ms=5, label='Men')
ax.plot(df_gen['year'], df_gen['women'], color='#E84855', lw=2.5, marker='s', ms=5, label='Women')
ax.fill_between(df_gen['year'], df_gen['men'], df_gen['women'], alpha=0.1, color='purple')
ax.set_title('Unemployment Rate by Gender (2010-2023)', fontsize=16, fontweight='bold')
ax.set_xlabel('Year'); ax.set_ylabel('Rate (%)')
ax.legend(); ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('../dashboard/chart3_gender.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Chart 4 — Regional Disparities

```python
df_s=df_reg.sort_values('rate_2023')
cols=['#E84855' if r>9 else '#FFA500' if r>7.5 else '#06D6A0' for r in df_s['rate_2023']]
fig,ax=plt.subplots(figsize=(12,7))
bars=ax.barh(df_s['region'], df_s['rate_2023'], color=cols, alpha=0.85)
for b,v in zip(bars,df_s['rate_2023']):
    ax.text(v+0.1, b.get_y()+b.get_height()/2, f'{v}%', va='center', fontsize=10, fontweight='bold')
ax.axvline(x=7.3, color='navy', ls='--', lw=1.5, label='National avg: 7.3%')
ax.set_title('Unemployment by Region — France 2023', fontsize=16, fontweight='bold')
ax.set_xlabel('Rate (%)'); ax.legend(); ax.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('../dashboard/chart4_regions.png', dpi=150, bbox_inches='tight')
plt.show()
```
